```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

# 한글 폰트 명시적으로 지정 (나눔고딕)
plt.rcParams['font.family'] = 'NanumGothic'
# matplotlib에서 음수 기호 표시 문제 해결
plt.rcParams['axes.unicode_minus'] = False

print(f"선택된 폰트: NanumGothic")

# CSV 경로
csv_path = r"C:\Users\USER\Desktop\FLIGHT_DATA\workspace\autoencoder\항공-날씨데이터 합친것(log).csv"

# 데이터 불러오기
try:
    df = pd.read_csv(csv_path, encoding='utf-8', on_bad_lines='skip')
except:
    try:
        df = pd.read_csv(csv_path, encoding='cp949', on_bad_lines='skip')
    except:
        df = pd.read_csv(csv_path, encoding='latin1', on_bad_lines='skip')

# 한글 컬럼 깨짐 방지를 위해 column 이름 출력
print("✅ 컬럼명:", df.columns.tolist())
print(df.head(3))  # 데이터 확인용

# 컬럼 순서대로 하나씩 출력
for col in df.columns:
    plt.figure(figsize=(10, 6))
    
    try:
        # 숫자형 데이터만 히스토그램 표시
        if pd.api.types.is_numeric_dtype(df[col]):
            sns.histplot(df[col].dropna(), bins=50, kde=True)
            plt.title(f"{col} 히스토그램")
            plt.xlabel(col)
            plt.ylabel("빈도수")
            print(f"✅ {col} 데이터 타입: {df[col].dtype}, 결측치: {df[col].isna().sum()}, 유니크 값 개수: {df[col].nunique()}")
        else:
            # 비숫자형 데이터는 막대 그래프로 표시 (상위 20개 값)
            value_counts = df[col].value_counts().nlargest(20)
            sns.barplot(x=value_counts.index, y=value_counts.values)
            plt.title(f"{col} 상위 20개 값")
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            print(f"✅ {col} 데이터 타입: {df[col].dtype}, 결측치: {df[col].isna().sum()}, 유니크 값 개수: {df[col].nunique()}")
    except Exception as e:
        plt.text(0.5, 0.5, f"{col} 시각화 실패\n오류: {e}", 
                horizontalalignment='center', verticalalignment='center',
                fontsize=12)
        plt.title(f"{col} (error)")
        print(f"⚠️ {col} 시각화 실패: {e}")
    
    plt.tight_layout()
    plt.show()
    
    # 기본 통계량 출력
    if pd.api.types.is_numeric_dtype(df[col]):
        print(f"{col} 기본 통계량:")
        print(df[col].describe())
    else:
        print(f"{col} 상위 10개 값:")
        print(df[col].value_counts().nlargest(10))
    
    print("\n" + "="*50 + "\n")  # 구분선
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

# 한글 폰트 명시적으로 지정 (나눔고딕)
plt.rcParams['font.family'] = 'NanumGothic'
# matplotlib에서 음수 기호 표시 문제 해결
plt.rcParams['axes.unicode_minus'] = False

print(f"선택된 폰트: NanumGothic")

# CSV 경로
csv_path = r"C:\Users\USER\Desktop\FLIGHT_DATA\workspace\all_flights_weather_data2.csv"

# 데이터 불러오기
try:
    df = pd.read_csv(csv_path, encoding='utf-8', on_bad_lines='skip')
except:
    try:
        df = pd.read_csv(csv_path, encoding='cp949', on_bad_lines='skip')
    except:
        df = pd.read_csv(csv_path, encoding='latin1', on_bad_lines='skip')

# 강수량 컬럼 확인 (컬럼명에 '강수량'이 포함된 컬럼 찾기)
rainfall_cols = [col for col in df.columns if '강수량' in col]
print("강수량 관련 컬럼:", rainfall_cols)

# 각 강수량 컬럼 분석
for col in rainfall_cols:
    # 최대값과 최소값 확인 및 출력
    min_val = df[col].min()
    max_val = df[col].max()
    print(f"\n{col} 최소값: {min_val}, 최대값: {max_val}")
    
    # 기본 통계량 출력
    print(f"\n{col} 기본 통계량:")
    print(df[col].describe())
    
    # 양수 데이터만 선택 (0 이상)
    positive_data = df[df[col] >= 0][col]
    
    # 히스토그램 그리기
    plt.figure(figsize=(12, 8))
    
    sns.histplot(positive_data, bins=50, kde=True)
    
    # x축 범위를 0부터 최대값까지 설정
    plt.xlim(0, max_val)
    
    plt.title(f"{col} 히스토그램 (0 ~ 최대값 {max_val})")
    plt.xlabel(col)
    plt.ylabel("빈도수")
    
    plt.tight_layout()
    plt.show()
    
    # 추가 정보 출력
    print("\n양수 데이터 통계:")
    print(positive_data.describe())
    
    # 0값 확인
    zero_count = (df[col] == 0).sum()
    print(f"0값 개수: {zero_count} ({zero_count/len(df)*100:.2f}%)")
    
    # 음수값 확인
    neg_count = (df[col] < 0).sum()
    print(f"음수값 개수: {neg_count} ({neg_count/len(df)*100:.2f}%)")
    
    # 결측치 확인
    na_count = df[col].isna().sum()
    print(f"결측치 개수: {na_count} ({na_count/len(df)*100:.2f}%)")
    
    print("\n" + "="*70 + "\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.patches import Rectangle
from matplotlib.patches import ConnectionPatch

# 한글 폰트 명시적으로 지정
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

# CSV 경로
csv_path = r"C:\Users\USER\Desktop\FLIGHT_DATA\workspace\all_flights_weather_data2.csv"

# 데이터 불러오기
try:
    df = pd.read_csv(csv_path, encoding='utf-8', on_bad_lines='skip')
except:
    try:
        df = pd.read_csv(csv_path, encoding='cp949', on_bad_lines='skip')
    except:
        df = pd.read_csv(csv_path, encoding='latin1', on_bad_lines='skip')

# 강수량 컬럼 찾기
rainfall_cols = [col for col in df.columns if '강수량' in col]
col = rainfall_cols[0]  # 첫 번째 강수량 컬럼 사용

# 양수 데이터만 선택 (0 이상)
positive_data = df[df[col] >= 0][col]

# 방법 1: 로그 스케일 그래프
plt.figure(figsize=(12, 6))
sns.histplot(positive_data, bins=50, kde=True)
plt.xscale('log')  # x축을 로그 스케일로 변환
plt.title(f"{col} 히스토그램 (로그 스케일)")
plt.xlabel(col)
plt.ylabel("빈도수")
# 0값이 있는 경우 log 스케일에서 표시되지 않으므로 주의 메시지 추가
if (positive_data == 0).any():
    plt.text(0.5, 0.9, "참고: 0값은 로그 스케일에서 표시되지 않습니다", 
            horizontalalignment='center', transform=plt.gca().transAxes)
plt.tight_layout()
plt.show()

# 방법 2: 브로큰 액시스 (Broken Axis) - 특정 범위를 건너뛰는 그래프
fig = plt.figure(figsize=(12, 6))

# 서브플롯 생성 (왼쪽: 0-100 범위, 오른쪽: 100-최대값 범위)
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

# 데이터 분리
low_range = positive_data[positive_data <= 100]
high_range = positive_data[positive_data > 100]

# 왼쪽 그래프 (0-100)
sns.histplot(low_range, bins=50, kde=True, ax=ax1)
ax1.set_xlim(0, 100)
ax1.set_title(f"{col} (0-100 범위)")
ax1.set_xlabel(col)
ax1.set_ylabel("빈도수")

# 오른쪽 그래프 (100-최대값)
if not high_range.empty:
    sns.histplot(high_range, bins=50, kde=True, ax=ax2)
    ax2.set_xlim(100, positive_data.max())
    ax2.set_title(f"{col} (100-{positive_data.max():.0f} 범위)")
    ax2.set_xlabel(col)
    ax2.set_ylabel("빈도수")
else:
    ax2.text(0.5, 0.5, "100 초과 값 없음", 
            horizontalalignment='center', verticalalignment='center')
    ax2.set_title(f"{col} (100 초과 범위)")

# 그래프 사이 간격 표시 (물결 모양)
d = .015  # 물결 크기
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((1-d, 1+d), (-d, +d), **kwargs)
ax1.plot((1-d, 1+d), (1-d, 1+d), **kwargs)

kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (-d, +d), **kwargs)
ax2.plot((-d, +d), (1-d, 1+d), **kwargs)

plt.tight_layout()
plt.show()

# 방법 3: 여러 범위로 나누어 그리기
plt.figure(figsize=(16, 12))

# 서브플롯 1: 0-10 범위
plt.subplot(2, 2, 1)
range1 = positive_data[positive_data <= 10]
sns.histplot(range1, bins=20, kde=True)
plt.title(f"{col} (0-10 범위)")
plt.xlabel(col)
plt.ylabel("빈도수")

# 서브플롯 2: 10-50 범위
plt.subplot(2, 2, 2)
range2 = positive_data[(positive_data > 10) & (positive_data <= 50)]
if not range2.empty:
    sns.histplot(range2, bins=20, kde=True)
    plt.title(f"{col} (10-50 범위)")
    plt.xlabel(col)
    plt.ylabel("빈도수")
else:
    plt.text(0.5, 0.5, "해당 범위 데이터 없음", 
            horizontalalignment='center', verticalalignment='center')
    plt.title(f"{col} (10-50 범위)")

# 서브플롯 3: 50-200 범위
plt.subplot(2, 2, 3)
range3 = positive_data[(positive_data > 50) & (positive_data <= 200)]
if not range3.empty:
    sns.histplot(range3, bins=20, kde=True)
    plt.title(f"{col} (50-200 범위)")
    plt.xlabel(col)
    plt.ylabel("빈도수")
else:
    plt.text(0.5, 0.5, "해당 범위 데이터 없음", 
            horizontalalignment='center', verticalalignment='center')
    plt.title(f"{col} (50-200 범위)")

# 서브플롯 4: 200 이상 범위
plt.subplot(2, 2, 4)
range4 = positive_data[positive_data > 200]
if not range4.empty:
    sns.histplot(range4, bins=20, kde=True)
    plt.title(f"{col} (200 이상 범위)")
    plt.xlabel(col)
    plt.ylabel("빈도수")
else:
    plt.text(0.5, 0.5, "해당 범위 데이터 없음", 
            horizontalalignment='center', verticalalignment='center')
    plt.title(f"{col} (200 이상 범위)")

plt.tight_layout()
plt.show()

# 추가 정보 출력
print(f"{col} 범위별 데이터 개수:")
print(f"0-10: {len(range1)} ({len(range1)/len(positive_data)*100:.2f}%)")
if 'range2' in locals():
    print(f"10-50: {len(range2)} ({len(range2)/len(positive_data)*100:.2f}%)")
if 'range3' in locals():
    print(f"50-200: {len(range3)} ({len(range3)/len(positive_data)*100:.2f}%)")
if 'range4' in locals():
    print(f"200 이상: {len(range4)} ({len(range4)/len(positive_data)*100:.2f}%)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 명시적으로 지정
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

# CSV 경로
csv_path = r"C:\Users\USER\Desktop\FLIGHT_DATA\workspace\all_flights_weather_data2.csv"

# 데이터 불러오기
try:
    df = pd.read_csv(csv_path, encoding='utf-8', on_bad_lines='skip')
except:
    try:
        df = pd.read_csv(csv_path, encoding='cp949', on_bad_lines='skip')
    except:
        df = pd.read_csv(csv_path, encoding='latin1', on_bad_lines='skip')

# 강수량 컬럼 찾기
rainfall_cols = [col for col in df.columns if '강수량' in col]
col = rainfall_cols[0]  # 첫 번째 강수량 컬럼 사용

# 양수 데이터만 선택 (0 이상)
positive_data = df[df[col] >= 0][col]

# 그래프 1: y축을 500까지 제한한 그래프
plt.figure(figsize=(12, 8))
ax = sns.histplot(positive_data, bins=50, kde=True)
plt.title(f"{col} 히스토그램 (y축 500까지 제한)")
plt.xlabel(col)
plt.ylabel("빈도수")
plt.ylim(0, 200)  # y축을 0-500으로 제한
plt.tight_layout()
plt.show()

# 그래프 2: 빈도수가 가장 많은 구간을 중심으로 그린 그래프
plt.figure(figsize=(12, 8))

# 히스토그램 데이터 계산
hist, bin_edges = np.histogram(positive_data, bins=50)
max_bin_idx = np.argmax(hist)  # 가장 빈도가 높은 빈의 인덱스

# 해당 빈의 범위 확인
max_bin_start = bin_edges[max_bin_idx]
max_bin_end = bin_edges[max_bin_idx + 1]

# 해당 범위 근처로 x축 제한 (앞뒤로 약간의 여유 추가)
x_min = max(0, max_bin_start - (max_bin_end - max_bin_start) * 3)
x_max = max_bin_end + (max_bin_end - max_bin_start) * 5

# 히스토그램 그리기
sns.histplot(positive_data, bins=50, kde=True)
plt.xlim(x_min, x_max)  # x축 범위 제한
plt.title(f"{col} 히스토그램 (빈도수 최대 구간 중심, x축 범위: {x_min:.1f}-{x_max:.1f})")
plt.xlabel(col)
plt.ylabel("빈도수")
plt.tight_layout()
plt.show()

# 추가 정보 출력
print(f"가장 빈도가 높은 구간: {max_bin_start:.1f} - {max_bin_end:.1f}")
print(f"해당 구간의 데이터 수: {hist[max_bin_idx]}")
print(f"전체 데이터 중 비율: {hist[max_bin_idx]/len(positive_data)*100:.2f}%")
```
