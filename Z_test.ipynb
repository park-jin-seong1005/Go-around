```python
from statsmodels.stats.proportion import proportions_ztest

# 변수별 데이터 (표에서 가져온 값)
data = {
    '강수량': {
        'High': (63, 3),
        'Medium': (84, 1),
        'Low': (5720, 55),
    },
    '기온': {
        'High': (1399, 24),
        'Medium': (3165, 28),
        'Low': (1303, 7),
    },
    '이슬점온도': {
        'High': (2269, 32),
        'Medium': (1270, 14),
        'Low': (2328, 13),
    },
    '해면기압': {
        'High': (2098, 11),
        'Medium': (2323, 19),
        'Low': (1446, 29),
    },
    '현지기압': {
        'High': (2283, 12),
        'Medium': (2695, 30),
        'Low': (889, 17),
    },
    '2분 평균풍속': {
        'High': (204, 10),
        'Medium': (1416, 25),
        'Low': (4247, 24),
    },
    '2분 최대풍속': {
        'High': (791, 28),
        'Medium': (1817, 16),
        'Low': (3259, 15),
    },
    '2분 최소풍속': {
        'High': (7, 0),
        'Medium': (548, 14),
        'Low': (5312, 45),
    }
}

# Z-test 수행 함수
def ztest_pairs(var_name, var_data):
    pairs = [('High', 'Low'), ('High', 'Medium'), ('Medium', 'Low')]
    print(f"\n📊 {var_name} Z-test 결과")
    for g1, g2 in pairs:
        if g1 in var_data and g2 in var_data:
            count = [var_data[g1][1], var_data[g2][1]]
            nobs = [var_data[g1][0], var_data[g2][0]]
            
            if sum(count) == 0:
                print(f"[{var_name}] {g1} vs {g2}: Go-around 횟수가 모두 0 → Z-test 불가")
                continue
            
            stat, pval = proportions_ztest(count, nobs)
            print(f"[{var_name}] {g1} vs {g2}: Z = {stat:.4f}, p = {pval:.10f}")

# 모든 변수에 대해 Z-test
for var, var_data in data.items():
    ztest_pairs(var, var_data)
```
